```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# -----------------------------------------------------------------------------
# Helper Functions for Logistic Regression
# -----------------------------------------------------------------------------

def sigmoid(z):
    """
    Compute the sigmoid activation function.

    Parameters:
        z: A scalar or np.array of any size.
        
    Returns:
        The sigmoid of z.
    """
    return 1 / (1 + np.exp(-z))

def compute_cost(X, y, weights, bias):
    """
    Compute the cross-entropy cost for logistic regression.

    Parameters:
        X       : NumPy array of features with shape (m, n)
        y       : NumPy array of labels with shape (m,)
        weights : NumPy array of weights (n,)
        bias    : Bias as a scalar
        
    Returns:
        cost    : The computed cost value.
    """
    m = X.shape[0]
    z = np.dot(X, weights) + bias
    predictions = sigmoid(z)
    epsilon = 1e-10  # to avoid log(0)
    cost = -np.sum(y * np.log(predictions + epsilon) +
                   (1 - y) * np.log(1 - predictions + epsilon)) / m
    return cost

def gradient_descent(X, y, weights, bias, learning_rate, num_iterations):
    """
    Perform batch gradient descent to learn weights and bias.

    Parameters:
        X             : Feature matrix (m x n)
        y             : Labels vector (m,)
        weights       : Initial weights (n,)
        bias          : Initial bias (scalar)
        learning_rate : Learning rate for updates
        num_iterations: Number of iterations
        
    Returns:
        weights      : Learned weights vector.
        bias         : Learned bias.
        cost_history : List of cost values for each iteration.
    """
    m = X.shape[0]
    cost_history = []

    for i in range(num_iterations):
        z = np.dot(X, weights) + bias
        predictions = sigmoid(z)
        
        # Calculate gradients.
        error = predictions - y
        grad_weights = np.dot(X.T, error) / m
        grad_bias = np.sum(error) / m
        
        # Parameter updates.
        weights -= learning_rate * grad_weights
        bias -= learning_rate * grad_bias
        
        # Record cost.
        cost = compute_cost(X, y, weights, bias)
        cost_history.append(cost)
        
    return weights, bias, cost_history

def plot_line(x_vals, y_vals, color, label):
    """
    Plot a line using the provided x and y values.
    
    Parameters:
        x_vals : List or array of x-values.
        y_vals : List or array of y-values.
        color  : Color of the line.
        label  : Label for the legend.
    """
    plt.plot(x_vals, y_vals, color=color, linewidth=2, label=label)

# -----------------------------------------------------------------------------
# Main Execution Function
# -----------------------------------------------------------------------------

def main():
    # Print current working directory (helps check file paths)
    print("Current working directory:", os.getcwd())
    
    # ------------------------------
    # Part 0: Data Loading and Preprocessing
    # ------------------------------
    # Update to the absolute path of your files.
    X_file_path = r"C:\Users\KIIT\Downloads\logisticX (1).csv"
    y_file_path = r"C:\Users\KIIT\Downloads\logisticY.csv"
    
    try:
        X_df = pd.read_csv(X_file_path, header=None)
    except FileNotFoundError:
        print(f"Error: File not found at: {X_file_path}")
        return

    try:
        y_df = pd.read_csv(y_file_path, header=None)
    except FileNotFoundError:
        print(f"Error: File not found at: {y_file_path}")
        return

    # Convert DataFrames to NumPy arrays.
    X = X_df.values  # Shape: (m, n)
    y = y_df.values.flatten()  # Shape: (m,)
    
    # Normalize features using z-score normalization.
    X_mean = np.mean(X, axis=0)
    X_std  = np.std(X, axis=0)
    X_norm = (X - X_mean) / X_std
    
    m, n = X_norm.shape
    print(f"Dataset loaded: {m} samples with {n} feature(s).")
    
    # ------------------------------
    # Part 1: Train Logistic Regression until Convergence (LR = 0.1)
    # ------------------------------
    learning_rate_conv = 0.1
    iterations_conv = 1000  # Sufficient iterations for convergence.
    weights_init = np.zeros(n)
    bias_init = 0.0
    
    weights_conv, bias_conv, cost_history_conv = gradient_descent(
        X_norm, y, weights_init, bias_init, learning_rate_conv, iterations_conv
    )
    
    print("\n--- Training Results (Learning Rate = 0.1) ---")
    print("Final Cost:", cost_history_conv[-1])
    print("Learned Weights:", weights_conv)
    print("Learned Bias:", bias_conv)
    
    # ------------------------------
    # Part 2: Plot Cost vs. Iteration (LR = 0.1)
    # ------------------------------
    plt.figure(figsize=(8, 6))
    plot_line(range(iterations_conv), cost_history_conv, "blue", "Cost (LR = 0.1)")
    plt.xlabel("Iteration")
    plt.ylabel("Cost")
    plt.title("Cost vs. Iteration (Learning Rate = 0.1)")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # ------------------------------
    # Part 3: Plot the Dataset and the Decision Boundary
    # (Only applicable for 2D data)
    # ------------------------------
    if n == 2:
        plt.figure(figsize=(8, 6))
        # Plot data points using plt.plot.
        plt.plot(X_norm[y == 0, 0], X_norm[y == 0, 1], "ro", label="Class 0")
        plt.plot(X_norm[y == 1, 0], X_norm[y == 1, 1], "bo", label="Class 1")
        
        # Calculate the decision boundary: weights[0]*x + weights[1]*y + bias = 0
        # => y = -(bias + weights[0]*x) / weights[1]
        x_vals = np.linspace(np.min(X_norm[:, 0]) - 0.5, np.max(X_norm[:, 0]) + 0.5, 100)
        if abs(weights_conv[1]) > 1e-7:
            y_vals = -(bias_conv + weights_conv[0] * x_vals) / weights_conv[1]
        else:
            y_vals = np.full_like(x_vals, np.mean(X_norm[:, 1]))
        plt.plot(x_vals, y_vals, "g-", linewidth=2, label="Decision Boundary")
        plt.xlabel("Feature 1 (Normalized)")
        plt.ylabel("Feature 2 (Normalized)")
        plt.title("Dataset with Decision Boundary")
        plt.legend()
        plt.grid(True)
        plt.show()
    else:
        print("Dataset is not 2-dimensional. Skipping decision boundary plot.")
    
    # ------------------------------
    # Part 4: Compare Two Learning Rates (0.1 vs 5) over 100 Iterations
    # ------------------------------
    iterations_short = 100
    init_weights = np.zeros(n)
    init_bias = 0.0
    
    # Train with learning rate 0.1.
    _, _, cost_history_lr_0_1 = gradient_descent(X_norm, y, init_weights, init_bias, 0.1, iterations_short)
    # Train with learning rate 5.
    _, _, cost_history_lr_5 = gradient_descent(X_norm, y, init_weights, init_bias, 5, iterations_short)
    
    plt.figure(figsize=(8, 6))
    plt.plot(range(iterations_short), cost_history_lr_0_1, "b-", linewidth=2, label="LR = 0.1")
    plt.plot(range(iterations_short), cost_history_lr_5, "m-", linewidth=2, label="LR = 5")
    plt.xlabel("Iteration")
    plt.ylabel("Cost")
    plt.title("Learning Rate Comparison (100 Iterations)")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # ------------------------------
    # Part 5: Compute Confusion Matrix and Evaluation Metrics
    # ------------------------------
    final_probs = sigmoid(np.dot(X_norm, weights_conv) + bias_conv)
    predictions = (final_probs >= 0.5).astype(int)
    
    TP = np.sum((predictions == 1) & (y == 1))
    TN = np.sum((predictions == 0) & (y == 0))
    FP = np.sum((predictions == 1) & (y == 0))
    FN = np.sum((predictions == 0) & (y == 1))
    
    conf_matrix = np.array([[TN, FP],
                            [FN, TP]])
    
    accuracy = (TP + TN) / m
    precision = TP / (TP + FP) if (TP + FP) != 0 else 0
    recall = TP / (TP + FN) if (TP + FN) != 0 else 0
    f1_score = (2 * precision * recall / (precision + recall)) if (precision + recall) != 0 else 0
    
    print("\n--- Evaluation Metrics ---")
    print("Confusion Matrix:")
    print(conf_matrix)
    print(f"Accuracy : {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall   : {recall:.4f}")
    print(f"F1 Score : {f1_score:.4f}")

if __name__ == "__main__":
    main()
```
